## Clean data

```python
# define a function to clean a loaded dataset

def clean(mydata):
    
    """
    This function cleans the input dataframe mydata:
    
    input:
        mydata: pandas.dataframe
    output: 
        pandas.dataframe

    """
    
    data = mydata.copy()
    
    #get rid of uneccessary columns in training and testing
    selected_cols = ['beer/ABV',
                     'beer/style',
                     'review/appearance',
                     'review/aroma',
                     'review/overall',
                     'review/palate',
                     'review/taste',
                     'user/ageInSeconds',
                     'user/gender']
    
    data = data[selected_cols]
    
    data['beer/style'] = pd.Categorical(data['beer/style'])
    data['beer/style_code'] = data['beer/style'].cat.codes
    
    data['user/gender'] = pd.Categorical(data['user/gender'])
    data['user/gender_code'] = data['user/gender'].cat.codes
    
    data['user/ageInSeconds'] = data['user/ageInSeconds'] / 60 / 60 / 24 / 365.25
    data['user/ageInSeconds'] = data['user/ageInSeconds'].round()

    #because there are some "bad lines" a.k.a lines that are shifted to the right due to bad parsing
    #subset rows that are correctly parsed
    data = data.dropna()
    
    return data
```

```python
print("Data dim before cleaning:", data.shape)
data = clean(data)
print("Data dim after cleaning:", data.shape)
```

```python
import numpy as np;
import seaborn as sns; sns.set()

def corrmatrix(mydata,annot=True):
        
    """
    This function cleans the input dataframe mydata:
    
    input:
        mydata: pandas.dataframe
    output: 
        pandas.dataframe

    """
    
    data = mydata.copy()
    
    data_continuos = data.select_dtypes(include=['int64','float64'])
    
    continuous_variables = list(data_continuos)
    
    corr = data_continuos.corr()
    
    mask = np.zeros_like(corr)
    mask[np.triu_indices_from(mask)] = True
    
    with sns.axes_style("white"):
        ax = sns.heatmap(corr, mask=mask, annot=annot, linewidths=.5)
```

## Clustering

```python
data['beer/style'].unique()
```

```python
import matplotlib.pyplot as plt

datac = data.copy()

selected_cols = ['review/appearance',
                 'review/aroma',
                 'review/overall',
                 'review/palate',
                 'review/taste',
                 'user/ageInSeconds',
                 'beer/style_code',
                 'user/gender_code']

datac=datac[selected_cols]

# k means clustering
from sklearn.cluster import KMeans
from sklearn import metrics
from scipy.spatial.distance import cdist

# first step is to determine k
distortions = []
score=[]
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(datac)
    kmeanModel.fit(datac)
    distortions.append(sum(np.min(cdist(datac, kmeanModel.cluster_centers_,
                                        'euclidean'),axis=1)) / datac.shape[0])
    score.append(kmeanModel.fit(datac).score(datac) )


#Plot the elbow
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
# plot centroids and clusters
from sklearn import cluster

def cluster_customer(n_clusters):
    centroids,labels,inertia = cluster.k_means(datac,n_clusters=n_clusters)
    #Let's check the parameter cluster centers of the estimator

    centroids = pd.DataFrame(centroids,
                             columns= [
                                 'review/appearance',
                                 'review/aroma',
                                 'review/overall',
                                 'review/palate',
                                 'review/taste',
                                 'user/ageInSeconds',
                                 'beer/style_code',
                                 'user/gender_code']
                            )
    
    sns.heatmap(centroids, cmap='BuPu', annot=True)
```
